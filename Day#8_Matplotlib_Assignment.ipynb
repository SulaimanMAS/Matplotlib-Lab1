```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
dict1 = {
    'x': [10 , 20 ,30,40],
    'y': [10 , 20 ,30 ,40]
}
x = [1,2,3,4]
y = [1,2,3,4]
plt.plot(x , y)
df = pd.DataFrame(dict1)
df.plot(x = 'x')
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(x , explode=[0 , 0 , 0 ,0])
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(x ,y, s = 50 , alpha=0.4)
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
df.plot(x="x")
plt.title("X and Y " , c='r'  ,loc='left')

plt.xlabel('X' , fontdict={'size':16})
plt.ylabel('Y' , fontdict={'size':16})
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot(range(0 , 40 , 10) , [40 , 30 , 20 , 10 ])
plt.plot(range(0 , 40 , 10) , [30 , 40 , 20 , 10 ])
plt.plot(range(0 , 40 , 10) , [10 , 40 , 20 , 30 ])
plt.legend(['X' , 'Y' , 'Z'] , loc = 'lower left')
```

## Q6: Create subplots with three charts and figsize=20,10

```python
fig , ax = plt.subplots(1 , 3 , figsize=(20,10))
ax[1].pie(x)
ax[2].plot(x , y)
ax[0].scatter(x , y )
```
